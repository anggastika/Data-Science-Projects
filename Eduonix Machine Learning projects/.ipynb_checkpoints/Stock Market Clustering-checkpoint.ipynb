```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import numpy as np
```

```python
companies_dict = {
'Amazon': 'AMZN',
'Apple': 'AAPL',
'Walgreen': 'WBA',
'Northrop Grumman': 'NOC',
'Boeing': 'BA',
'Lockheed Martin': 'LMT',
'McDonalds': 'MCD',
'Intel': 'INTC',
'Navistar': 'NAV',  
'IBM': 'IBM',
'Texas Instruments': 'TXN',
'MasterCard': 'MA',
'Microsoft': 'MSFT',
'General Electrics': 'GE',
'Symantec': 'SYMC',
'American Express': 'AXP',
'Pepsi': 'PEP',
'Coca Cola': 'KO',
'Johnson & Johnson': 'JNJ',
'Toyota': 'TM',
'Honda': 'HMC',
'Sony': 'SNE',
'Exxon': 'XOM',
'Chevron': 'CVX',
'Valero Energy': 'VLO',
'Ford': 'F',
'Bank of America': 'BAC'}

#sort the dictionary by stock labels
companies = sorted(companies_dict.items(), key=lambda x:x[1])
#define the datasource to use
data_source = "iex"
#define the start and end dates
start_date = "2017-12-31"
end_date = "2019-03-29"
#use pandas_reader.data.DataReader
data_list = []
for item in companies_dict.values():
    data_list.append(data.DataReader(item, data_source, start_date, end_date))
```

```python
#calculating stock movement
for d in data_list:
    d["movement"] = d["close"] - d["open"]
data_list[0]
```

```python
#Put all of the data into a matrix
X = []
for d in data_list:
    X.append(d["movement"])
X = np.array(X)
X
```
