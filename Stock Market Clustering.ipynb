# Stock Movement Clustering

Obtained open and closing values of various stocks using pandas_datareader. Calculated the daily movements for ~400 days, then performed kmeans clustering for 5 cluster centers. Then used principal component analysis to reduce the dimensionality of the data into only 2 dimensions. The clusters were very interesting and made a lots of sense. Clusters consisted of similar companies. For example, car companies were in the same cluster. Other clusters consisted of airplane companies, tech companies, or food and personal product companies.

```python
#import the necessary libraries. Pandas_datareader is used to access the stock market API
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import numpy as np
```

Define a dictionary of companies and their tick markers

```python
companies_dict = {
'Amazon': 'AMZN',
'Apple': 'AAPL',
'Walgreen': 'WBA',
'Northrop Grumman': 'NOC',
'Boeing': 'BA',
'Lockheed Martin': 'LMT',
'McDonalds': 'MCD',
'Intel': 'INTC',
'Navistar': 'NAV',  
'IBM': 'IBM',
'Texas Instruments': 'TXN',
'MasterCard': 'MA',
'Microsoft': 'MSFT',
'General Electrics': 'GE',
'Symantec': 'SYMC',
'American Express': 'AXP',
'Pepsi': 'PEP',
'Coca Cola': 'KO',
'Johnson & Johnson': 'JNJ',
'Toyota': 'TM',
'Honda': 'HMC',
'Sony': 'SNE',
'Exxon': 'XOM',
'Chevron': 'CVX',
'Valero Energy': 'VLO',
'Ford': 'F',
'Bank of America': 'BAC'}

#keep the names in an array
companies = list(companies_dict.keys())
#define the datasource to use
data_source = "iex"
#define the start and end dates
start_date = "2017-12-31"
end_date = "2019-03-29"
#use pandas_reader.data.DataReader
#the function keeps giving me errors, my way to get around it is to read them one at a time.
data_list = []
for item in companies_dict.values():
    data_list.append(data.DataReader(item, data_source, start_date, end_date))
```

```python
#calculating stock movement
for d in data_list:
    d["movement"] = d["close"] - d["open"]
data_list[0]
```

Visualizing data

```python
#plotting the movement of a stock
for i,d in enumerate(data_list):
    plt.plot(d["movement"])
    plt.xticks([10,100,200,300])
    plt.title(list(companies_dict.keys())[i])
    plt.ylabel("Movement")
    plt.xlabel("Date")
    plt.show()
```

Setting up training set

```python
#Put all of the data into a matrix
#columns for the sample and rows for each day's movement
X = []
for d in data_list:
    X.append(d["movement"])
X = np.array(X)
X
```

```python
#building a pipeline
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans
from sklearn.preprocessing import Normalizer
norm = Normalizer()
km = KMeans(n_clusters =5, max_iter=1000)
pipe = make_pipeline(norm,km)
#fitting data
pipe.fit(X)
print("error: {}".format(km.inertia_))
```

Make a dataframe to store all the data

```python
clusters = pd.DataFrame({"label":km.labels_,"companies":companies})
```

It's interesting to see similarities between McDonald's and soda companies, between car companies, and plane companies. That tells us that the stock price movements are connected between similar companies. 

```python
#Visualize the data using PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
reduced = pca.fit_transform(norm.fit_transform(X))
clusters['c1'] = reduced[:,0] #first principal component
clusters['c2'] = reduced[:,1] #second principal component
for c in sorted(clusters['label'].value_counts().index):
    this_cluster = clusters[clusters['label']==c]
    print("cluster:{}, \ncompanies:\n{}\n".format(c,this_cluster['companies']))
    plt.scatter(this_cluster['c1'],this_cluster['c2'])

plt.legend(labels=sorted(clusters['label'].value_counts().index))
plt.title('Reduced to 2 dimensions using PCA')
plt.show()
```
