# Image denoising using PCA

PCA have been used for dimensionality reduction to increase training speed for machine learning models. Another interesting application is image denoising.


```python
%matplotlib inline
#importing the necessary libraries
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import copy
#read an image
img = mpimg.imread("cameraman.tif")
img = img.astype(float)
#plot image
plt.imshow(img,cmap="Greys_r")
print(img.shape)
```

Adding gaussian noise into the image

```python
noise = np.random.normal(-20,20,img.shape)
plt.imshow(img+noise,cmap="Greys_r")
```

Making a denoise function. The idea is find a center window within the image, then grab a bigger window around the center window as the training set. Within each training window, have a moving window the same size as the center window in the training window. Then flatten out each of the sub images to form a matrix X. Using kmeans, you can find the rows that are most similar to the center window. Take only the similar rows and perform singular value decomposition (training the PCA algorithm). This will obtain the singular matrixes from X. Then we can use the truncated left singular matrix to project the center window to a lower dimension space. Then reconstruct the projected matrix back to the original space. Because the signal-to-noise ratio in the first few principal components are higher, projecting with a truncated singular matrix causes a reduction of noise level.

The function below takes an image of a certain size, construct a moving window of k x k dimension, which is the same size as the center window we will use. Flatten each window and point on a row in X matrix. This will be the training set for PCA.

```python
def build_training_set(sub_img, k): 
    #take an image or a sub image, define a kxk moving window
    #flatten all of the windows to form a training set
    h,w = sub_img.shape
    X = []
    for i in range(0,h): #vertical direction
        if i > k: #ignore the first k-1 data points
            for j in range(0,w): #horizonal direction
                if j > k: #ignore the first k-1 data points
                    X.append(np.ravel(sub_img[i-k:i,j-k:j])) # a window of k x k
    return np.array(X)
```

This function use kmeans to find the most similar moving windows to form X, which is used to train PCA. Then use the left singular matrix obtained to project and reconstruct the center window. The expected outcome is that, during the transformations, noise in each center window is reduced.

```python
#import the necessary libraries
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
#
def train_predict(X,center_in,N,pca_n):
    #X is an nxm matrix
    #center_in is 1Xm matrix
    #N is the number times kmeans is run
    #pca_n is the n_components of the PCA
    center = [np.ravel(center_in)] #flatten the center window
    assert(len(center[0])==len(X[0]))
    #feature scaling
    SS = StandardScaler()
    X = SS.fit_transform(X)
    center = SS.transform(center)
    #divide the data into 2 clusters for a number of times.
    #In each iteration, retain only the cluster most similar to the center window
    for n in range(0,N):
        pca_X = []
        km = KMeans(n_clusters=2)
        km.fit(X)
        prediction = km.predict(center)
        i = 0
        for l,x in zip(km.labels_,X): #append data belonging to the center window cluster
            if l == prediction:
                pca_X.append(x)
            i += 1
        X = np.array(pca_X)
    #Principal component analysis
    pca = PCA(n_components=pca_n)
    pca.fit(np.array(pca_X)) #calculate the singular matrixes
    projected = pca.transform(center) #project the center data onto the principal 
    recons = pca.inverse_transform(projected) #transform back to the original space
    recons = SS.inverse_transform(recons) #reverse the feature scaling
    return recons[0].reshape(center_in.shape) #reshape the flatten array back to the center window
```

This splits the whole image into smaller pieces of equal size. These are the center windows. It then takes a training window of L x L from surrounding pixels that encloses the center window. The training and center window are then sent into build_training_set and train_predict to obtain the predicted center window to replace the original center window.

```python
def denoise(img_in,kh,kw,L,km_N,pca_n):
    #img_in: original image
    #kh: height of center window
    #kw: width of center window
    #L: dimension of training window
    #km_N: number of times kmeans of 2 clusters is run
    #pca_n: n_components of PCA
    img = copy.copy(img_in)
    h,w = img.shape
    assert(kh < L and kw < L)
    assert(L%2 == 0 and kh%2 == 0 and kw%2 ==0)
    assert(h%kh == 0 and w%kw == 0)
    for i in range(0,h//kh): #h//kh: divide the image height by the center window height
        for j in range(0,w//kw): #w//kw: divide the image width by the center window width
            c_i,c_j = i*kh+kh//2,j*kw+kw//2 #c_i and c_j are the center indexes
            center = img[i*kh:i*kh+kh,j*kw:j*kw+kw] #center window
            #Below are the corner cases if c_i and c_j are less then L//2 from the edges
            #h1, h2, w1, w2: col_start, col_end, row_start, row_end
            if c_i < L//2: #upper bound
                h1,h2 = 0,L
            else:
                if h-c_i < L//2: #lower bound
                    h1,h2 = h-L,h
                else:
                    h1,h2 = c_i-L//2,c_i+L//2
            if c_j < L//2: #left bound
                w1,w2 = 0,L
            else:
                if w-c_j < L//2: #right bound
                    w1,w2 = w-L,w
                else:
                    w1,w2 = c_j-L//2,c_j+L//2 #normal operations
            train = img[h1:h2,w1:w2] #get training set
            X = build_training_set(train,kh) #call function (use kh as k)
            img[i*kh:i*kh+kh,j*kw:j*kw+kw] = train_predict(X,center,km_N,pca_n) #replace the center window by the predict
    return img
```

Add noise into the original image and then send into the denoise function.

```python
image = denoise(img+noise,16,16,26,3,0.5)
plt.imshow(image,cmap="Greys_r")
```
